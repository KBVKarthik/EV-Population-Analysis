6. Clustering and Segmentation
Clustering EVs: Use clustering techniques (e.g., K-means) to group vehicles based on features like electric range, MSRP, and type to identify distinct market segments.
Legislative District Segmentation: Segment districts based on EV adoption patterns, eligibility, and vehicle types.

**ai_message:**

To perform clustering and segmentation, I will first preprocess the data to ensure it is suitable for clustering (e.g., scaling numeric features) and then apply K-means clustering to group vehicles based on electric range, MSRP, and type. For legislative district segmentation, I will analyze patterns in EV adoption, eligibility, and vehicle types. Let's proceed.

```
# Clustering EVs using K-means
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler

# Selecting relevant features for clustering
features = df[['Electric Range', 'Base MSRP', 'Electric Vehicle Type']]

# Encoding categorical variable 'Electric Vehicle Type'
features['Electric Vehicle Type'] = features['Electric Vehicle Type'].astype('category').cat.codes

# Scaling the features
scaler = StandardScaler()
scaled_features = scaler.fit_transform(features)

# Applying K-means clustering
kmeans = KMeans(n_clusters=3, random_state=42)
df['Cluster'] = kmeans.fit_predict(scaled_features)

# Visualizing the clusters
plt.figure(figsize=(10, 6))
sns.scatterplot(data=df, x='Electric Range', y='Base MSRP', hue='Cluster', palette='viridis')
plt.title('K-means Clustering of EVs')
plt.xlabel('Electric Range (miles)')
plt.ylabel('Base MSRP ($)')
plt.grid(True)
plt.show()

# Cluster centers
print("Cluster Centers:")
print(kmeans.cluster_centers_)
```

**ai_message:**

Let me complete the legislative district segmentation analysis:

```
# Legislative District Segmentation
district_features = df.groupby('Legislative District').agg({
    'Electric Range': 'mean',
    'Base MSRP': 'mean',
    'CAFV Eligibility Simple': lambda x: (x == 'Eligible').mean()
}).fillna(0)

# Scaling district features
scaler = StandardScaler()
scaled_district = scaler.fit_transform(district_features)

# Applying K-means clustering
kmeans_district = KMeans(n_clusters=3, random_state=42)
district_features['Cluster'] = kmeans_district.fit_predict(scaled_district)

# Visualizing district segments
plt.figure(figsize=(10, 6))
sns.scatterplot(data=district_features, x='Electric Range', y='Base MSRP', hue='Cluster', palette='deep')
plt.title('Legislative District Segments')
plt.xlabel('Average Electric Range')
plt.ylabel('Average Base MSRP')
plt.grid(True)
plt.show()

print("\nDistrict Cluster Characteristics:")
print(district_features.groupby('Cluster').mean())
```

7. Predictive Modeling
Predicting Electric Range: Build a regression model to predict electric range based on factors like MSRP, model year, and type.
CAFV Eligibility Prediction: Use classification models to predict whether a vehicle is eligible for CAFV incentives based on its features.

**ai_message:**

To predict electric range and CAFV eligibility, I will preprocess the data, build a regression model for electric range prediction, and a classification model for CAFV eligibility. Let's proceed with the analysis.

```
# Predicting Electric Range using Regression
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score

# Selecting features and target for regression
regression_features = df[['Base MSRP', 'Model Year', 'Electric Vehicle Type']]
regression_features['Electric Vehicle Type'] = regression_features['Electric Vehicle Type'].astype('category').cat.codes
regression_target = df['Electric Range']

# Splitting the data
X_train, X_test, y_train, y_test = train_test_split(regression_features, regression_target, test_size=0.2, random_state=42)

# Building the regression model
reg_model = LinearRegression()
reg_model.fit(X_train, y_train)

# Predictions and evaluation
y_pred = reg_model.predict(X_test)
rmse = mean_squared_error(y_test, y_pred, squared=False)
r2 = r2_score(y_test, y_pred)

print("Regression Model Evaluation:")
print("RMSE:", rmse)
print("R-squared:", r2)
```

```
# CAFV Eligibility Classification
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import classification_report

# Preparing features for classification
classification_features = df[['Electric Range', 'Base MSRP', 'Electric Vehicle Type']]
classification_features['Electric Vehicle Type'] = classification_features['Electric Vehicle Type'].astype('category').cat.codes
classification_target = df['CAFV Eligibility Simple'].map({'Eligible': 1, 'Not Eligible': 0})

# Splitting data
X_train, X_test, y_train, y_test = train_test_split(classification_features, classification_target, test_size=0.2, random_state=42)

# Training Random Forest
rf_model = RandomForestClassifier(random_state=42)
rf_model.fit(X_train, y_train)

# Predictions and evaluation
y_pred = rf_model.predict(X_test)

# Feature importance visualization
plt.figure(figsize=(10, 5))
importance = pd.DataFrame({
    'Feature': classification_features.columns,
    'Importance': rf_model.feature_importances_
}).sort_values('Importance', ascending=False)

sns.barplot(data=importance, x='Feature', y='Importance')
plt.title('Feature Importance for CAFV Eligibility Prediction')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

print("\nClassification Report:")
print(classification_report(y_test, y_pred))
```
